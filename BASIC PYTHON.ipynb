## **THEORY**

1. What is Python, and why is it popular?
- Python is a high-level, interpreted programming language known for its simple and readable syntax. It’s popular because it's easy to learn, supports multiple programming paradigms, and has a large community with tons of libraries and frameworks for everything from web development to data science.

2. What is an interpreter in Python?
- An interpreter is a program that reads and executes Python code line by line. It converts the Python code into machine code at runtime, which allows for immediate feedback and easier debugging.

3. What are pre-defined keywords in Python?
- Pre-defined keywords in Python are reserved words that have special meaning in the language. For example, if, for, while, True, False, etc. These cannot be used as variable names.

4. Can keywords be used as variable names?
- No, keywords cannot be used as variable names because they are reserved for specific language syntax and logic.

5. What is mutability in Python?
- Mutability refers to whether an object’s content can be changed after it's created. If it can be changed, it’s called mutable. If not, it’s immutable.

6. Why are lists mutable, but tuples are immutable?
- Lists are designed to store collections that may need to change — like adding or removing items. Tuples, on the other hand, are used for fixed collections, so Python makes them immutable to prevent accidental changes and to allow them to be used as dictionary keys or set elements.

7. What is the difference between “==” and “is” operators in Python?
- ==checks if two values are equal, while is checks if two variables refer to the same object in memory.

8. What are logical operators in Python?
- Logical operators are used to combine conditional statements. The main ones are:

  and (True if both are True)

  or (True if at least one is True)

  not (Reverses the result)

9. What is type casting in Python?
- Type casting is converting a variable from one data type to another, like from string to int, or float to int.

10. What is the difference between implicit and explicit type casting?
- Implicit casting is done automatically by Python, like when adding an integer and a float.

  Explicit casting is done manually using functions like int(), float(), str(), etc.

11. What is the purpose of conditional statements in Python?
- Conditional statements are used to execute certain blocks of code only when specific conditions are true. They help control the flow of the program.

12. How does the elif statement work?
- elif stands for "else if." It lets you check multiple conditions one after  another, after the initial if condition. Only the first true condition runs.

13. What is the difference between for and while loops?
- for loops are used when you know in advance how many times to loop.

  while loops are used when the loop should continue until a condition becomes false.

14. Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is better when you don't know how many times you'll need to loop
 — like asking a user to enter a password until they get it right. You can’t predict how many attempts they'll need.

```python
#4. Write a program that checks if a given word is a Python keyword.

import keyword
word = input("Enter a word to check if it's a Python keyword: ")

if keyword.iskeyword(word):
    print(f"Yes, '{word}' is a Python keyword.")
else:
    print(f"No, '{word}' is not a Python keyword.")
```

```python
#6. Write a function to demonstrate the behavior of mutable and immutable arguments.
def change_values(my_list, my_number):
    print("Before changes:")
    print("List:", my_list)
    print("Number:", my_number)

    my_list.append(100)
    my_number += 50

    print("After changes:")
    print("List:", my_list)
    print("Number:", my_number)

my_list = [1, 2, 3]
my_number = 10

change_values(my_list, my_number)

print("\nOutside the function:")
print("List:", my_list)
print("Number:", my_number)
```

```python
#10. Write code to demonstrate type casting with list elements.
i_list= [1, 2.5, '3', True]

int_list = []
float_list = []
str_list = []
bool_list = []

for item in i_list:
    int_list.append(int(item))

for item in i_list:
    float_list.append(float(item))

for item in i_list:
    str_list.append(str(item))

for item in i_list:
    bool_list.append(bool(item))

print("Original List:", i_list)
print("Integer List:", i_list)
print("Float List:", i_list)
print("String List:", i_list)
print("Boolean List:", i_list)
```

```python
#14. Write a program to reverse a string using a while loop.
input_string = input("Enter a string: ")
reversed_string = ""
index = len(input_string) - 1

while index >= 0:
    reversed_string += input_string[index]
    index -= 1
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
number = int(input("Enter a number: "))
factorial = 1
current = 1

while current <= number:
    factorial *= current
    current += 1

print(f"The factorial of {number} is {factorial}")
```
